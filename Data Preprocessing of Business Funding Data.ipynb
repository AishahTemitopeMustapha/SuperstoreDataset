### 2. Handling Missing Values

Explanation:
Missing values can impact the quality of data analysis. Analyzing the percentage of missing values helps in making informed decisions. If a column has too many missing values, it may be dropped. For other missing values, we apply appropriate strategies such as filling them with default values.



```python
# Check for missing values
missing_values = missing_percentage = (df.isnull().sum() / len(df)) * 100
print("\nMissing Values:\n", missing_values[missing_values > 0])

# Drop 'Effective date' column due to 76% missing values
df.drop(columns=['Effective date'], inplace=True)
print("\nDropped 'Effective date' column due to high missing values.")

# Fill missing 'Financing Type Normalized' with "Unknown"
df["Financing Type Normalized"].fillna("Unknown", inplace=True)

# Fill missing values in 'Investors Count' with 0 (or use median)
df["Investors Count"].fillna(0, inplace=True)
```

### 3. Removing Duplicates

Explanation:
Duplicates can cause bias in data analysis. We remove duplicate records to ensure data integrity.


```python
# Drop duplicates
df.drop_duplicates(inplace=True)
print("\nDuplicates Removed. New Shape:", df.shape)
```

### 4. Transforming Date Columns

Explanation:
Dates provide valuable insights. We convert the 'Found At' column into a proper datetime format and extract the year, month, and day for further analysis.


```python
# Convert 'Found At' column to datetime format with UTC
df['Found At'] = pd.to_datetime(df['Found At'], errors='coerce', utc=True)

# Drop rows where conversion failed (optional)
df = df.dropna(subset=['Found At'])

# Extract Year, Month, and Day
df['Year'] = df['Found At'].dt.year
df['Month'] = df['Found At'].dt.month
df['Day'] = df['Found At'].dt.day

# Drop 'Found At' column since it's now redundant
df.drop(columns=['Found At'], inplace=True)
print("\nDropped 'Found At' column as Year, Month, and Day have been extracted.")

print("\nSuccessfully separated 'Found At' into 'Year', 'Month', and 'Day' columns.")

print("\nSeparated 'Found At' into 'Year', 'Month', and 'Day' columns.")
```

### 5. Dropping Redundant Columns

Explanation:
Some columns contain duplicate information in different formats. We drop 'Amount' and 'Investors' as their normalized versions exist.


```python
# Remove unnormalized columns ('Amount', 'Investors' 'Financing Type') since normalized versions exist
df.drop(columns=['Amount', 'Investors', 'Financing Type'], inplace=True)
print("\nDropped 'Amount' , 'Investors' and 'Financing Type' columns since normalized versions exist.")
```
